#Aumentando as probabilidades

```python
import random

def sorteia_mega_sena():
  """Sorteia 6 nmeros aleatrios para a Mega-Sena"""
  numeros_sorteados = random.sample(range(1, 61), 6)
  return sorted(numeros_sorteados)

# Exemplo de uso:
resultado = sorteia_mega_sena()
print(resultado)
```

```python
from scipy.special import comb

def probabilidade(numeros_jogados, numeros_acertar):
  """Calcula a probabilidade de acertar x nmeros em um jogo da Mega-Sena"""
  total_combinacoes = comb(60, 6)  # Total de combinaes possveis na Mega-Sena
  combinacoes_favoraveis = comb(numeros_jogados, numeros_acertar) * comb(60 - numeros_jogados, 6 - numeros_acertar)
  probabilidade = combinacoes_favoraveis / total_combinacoes
  return probabilidade

# Exemplo: Probabilidade de acertar 4 nmeros jogando 7
probabilidade_4_numeros = probabilidade(6, 4)
print(probabilidade_4_numeros)
```

```python
numeros_jogados = range(6, 16) #Essa parte cria uma sequncia de nmeros que comea em 6 (inclusive) e vai at 15 (exclusivo). Ou seja, a sequncia gerada ser: 6, 7, 8, 9, 10, 11, 12, 13, 14.
probabilidades_sena = []
probabilidades_quina = []
probabilidades_quadra = []

for n in numeros_jogados:
  probabilidades_sena.append(probabilidade(n, 6)) #Sena #append permite que voc adicione um elemento ao final de uma lista existente.
  probabilidades_quina.append(probabilidade(n, 5))#Quina
  probabilidades_quadra.append(probabilidade(n, 4))#Quadra

plt.plot(numeros_jogados, probabilidades_sena, label='Sena')
plt.plot(numeros_jogados, probabilidades_quina, label='Quina')
plt.plot(numeros_jogados, probabilidades_quadra, label='Quadra')
plt.xlabel('Nmero de nmeros jogados')
plt.ylabel('Probabilidade')
plt.title('Probabilidade de ganhar na Mega-Sena')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados (supondo que os dados estejam em um arquivo CSV)
df = pd.read_csv('Mega-Sena_cleanAllLoser.csv')

# Contar a frequncia de cada nmero
contagem_numeros = df.stack().value_counts()

# Plotar um grfico de barras
plt.figure(figsize=(12, 6))
contagem_numeros.plot(kind='bar')
plt.title('Frequncia dos nmeros da Mega-Sena')
plt.xlabel('Nmero')
plt.ylabel('Frequncia')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assumindo que seus dados esto em um DataFrame chamado 'df'
# e que as colunas com os nmeros sorteados so 'Bola1', 'Bola2', ..., 'Bola6'

# Configurando o estilo do grfico (opcional)
plt.style.use('ggplot')

# Criando o grfico de disperso
plt.figure(figsize=(12, 6))

# Plotando cada nmero sorteado como um ponto azul
for i in range(1, 7):  # Assumindo que voc tem 6 colunas de nmeros
    plt.scatter(df.index, df[f'Bola{i}'], color='red', label=f'Bola {i}')

plt.xlabel('Concurso')
plt.ylabel('Nmero Sorteado')
plt.title('Distribuio dos Nmeros Sorteados na Mega-Sena')
plt.legend()
plt.grid(True)
plt.show()
```
