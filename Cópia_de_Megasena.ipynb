#Aumentando as probabilidades

```python
import random

def sorteia_mega_sena():
  """Sorteia 6 números aleatórios para a Mega-Sena"""
  numeros_sorteados = random.sample(range(1, 61), 6)
  return sorted(numeros_sorteados)

# Exemplo de uso:
resultado = sorteia_mega_sena()
print(resultado)
```

```python
from scipy.special import comb

def probabilidade(numeros_jogados, numeros_acertar):
  """Calcula a probabilidade de acertar x números em um jogo da Mega-Sena"""
  total_combinacoes = comb(60, 6)  # Total de combinações possíveis na Mega-Sena
  combinacoes_favoraveis = comb(numeros_jogados, numeros_acertar) * comb(60 - numeros_jogados, 6 - numeros_acertar)
  probabilidade = combinacoes_favoraveis / total_combinacoes
  return probabilidade

# Exemplo: Probabilidade de acertar 4 números jogando 7
probabilidade_4_numeros = probabilidade(6, 4)
print(probabilidade_4_numeros)
```

```python
numeros_jogados = range(6, 16) #Essa parte cria uma sequência de números que começa em 6 (inclusive) e vai até 15 (exclusivo). Ou seja, a sequência gerada será: 6, 7, 8, 9, 10, 11, 12, 13, 14.
probabilidades_sena = []
probabilidades_quina = []
probabilidades_quadra = []

for n in numeros_jogados:
  probabilidades_sena.append(probabilidade(n, 6)) #Sena #append permite que você adicione um elemento ao final de uma lista existente.
  probabilidades_quina.append(probabilidade(n, 5))#Quina
  probabilidades_quadra.append(probabilidade(n, 4))#Quadra

plt.plot(numeros_jogados, probabilidades_sena, label='Sena')
plt.plot(numeros_jogados, probabilidades_quina, label='Quina')
plt.plot(numeros_jogados, probabilidades_quadra, label='Quadra')
plt.xlabel('Número de números jogados')
plt.ylabel('Probabilidade')
plt.title('Probabilidade de ganhar na Mega-Sena')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados (supondo que os dados estejam em um arquivo CSV)
df = pd.read_csv('Mega-Sena_cleanAllLoser.csv')

# Contar a frequência de cada número
contagem_numeros = df.stack().value_counts()

# Plotar um gráfico de barras
plt.figure(figsize=(12, 6))
contagem_numeros.plot(kind='bar')
plt.title('Frequência dos números da Mega-Sena')
plt.xlabel('Número')
plt.ylabel('Frequência')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assumindo que seus dados estão em um DataFrame chamado 'df'
# e que as colunas com os números sorteados são 'Bola1', 'Bola2', ..., 'Bola6'

# Configurando o estilo do gráfico (opcional)
plt.style.use('ggplot')

# Criando o gráfico de dispersão
plt.figure(figsize=(12, 6))

# Plotando cada número sorteado como um ponto azul
for i in range(1, 7):  # Assumindo que você tem 6 colunas de números
    plt.scatter(df.index, df[f'Bola{i}'], color='red', label=f'Bola {i}')

plt.xlabel('Concurso')
plt.ylabel('Número Sorteado')
plt.title('Distribuição dos Números Sorteados na Mega-Sena')
plt.legend()
plt.grid(True)
plt.show()
```
