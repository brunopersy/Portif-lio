**QuestÃ£o 1. Como os preÃ§os das propriedades mudaram ao longo do tempo?**

**Convertendo datas**

```python
pd.to_datetime(houses['date'])
```

**Salvando conversÃ£o de datas na variÃ¡vel (houses)**

```python
houses['date'] = pd.to_datetime(houses['date'])
```

```python
_ = avg_price_per_date = houses.groupby(['date'])['price'].mean()
_ = plt.plot(avg_price_per_date.index, avg_price_per_date.values)
```

Agora vamos add uma linha de tendÃªncia no grÃ¡fico.

```python
avg_price_per_date = houses.groupby(['date'])['price'].mean()
_ = plt.title('PreÃ§os das casas ao longo do tempo')


x = avg_price_per_date.index.values.astype(np.int64) // 10**9
y = avg_price_per_date.values


coefficients = np.polyfit(x, y, 1)  #lembrei dos polinÃ´mios de 1 grau
m = coefficients[0]
b = coefficients[1]


trendline = m * x + b


plt.plot(avg_price_per_date.index, avg_price_per_date.values, label='MÃ©dia de PreÃ§os')
plt.plot(avg_price_per_date.index, trendline, color='blue', label='Linha de TendÃªncia')


plt.xlabel('Data')
plt.ylabel('PreÃ§o MÃ©dio($)')
plt.title('PreÃ§os MÃ©dios de Casas por Data')
plt.legend()

plt.show()
```

**A linha de tendÃªncia, inclinada para cima, indica que houve uma tendÃªncia geral de aumento nos preÃ§os mÃ©dios das casas ao longo do perÃ­odo analisado.**

Apesar da tendÃªncia geral de alta, podemos observar algumas variaÃ§Ãµes sazonais nos preÃ§os. Em alguns perÃ­odos do ano, os preÃ§os mÃ©dios parecem ser mais altos ou mais baixos do que a mÃ©dia geral, indicando que fatores sazonais podem influenciar o mercado imobiliÃ¡rio.

Os preÃ§os mÃ©dios nÃ£o seguem uma trajetÃ³ria perfeitamente linear. Existem flutuaÃ§Ãµes e perÃ­odos de maior volatilidade, indicando que o mercado imobiliÃ¡rio pode ser suscetÃ­vel a mudanÃ§as e incertezas.

QuestÃ£o 3. Quais as regiÃµes mais caras?

Vamos investigar quais sÃ£o as 10 cidades mais caras.

```python
firist_top10 = houses.groupby('city')['price_in_k'].mean().sort_values(ascending=False).head(10)
firist_top10_cities = houses[houses['city'].isin(firist_top10.index)]
firist_top10_cities.groupby('city')['price'].count()
```

**Entretanto, serÃ¡ que essas cidades tem um nÃºmero relevante de imÃ³veis vendidos ou tem alguns poucos e esses sÃ£o outiliers.**

```python
cities_count = houses.groupby('city')['price'].count()
cities_over_30 = cities_count[cities_count > 30] #lista de cidades com amostra suficiente
cities_over_30
```

```python
houses[houses['city'].isin(cities_over_30)].groupby('city')['price'].mean().sort_values(ascending=False).head(10)
```
