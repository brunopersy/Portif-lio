#Tema Macro: Tendências de Mercado e Analise de Preços.
**Questão 1. Como os preços das propriedades mudaram ao longo do tempo?**

**Convertendo datas**

```python
pd.to_datetime(houses['date'])
```

**Salvando conversão de datas na variável (houses)**

```python
houses['date'] = pd.to_datetime(houses['date'])
```

```python
_ = avg_price_per_date = houses.groupby(['date'])['price'].mean()
_ = plt.plot(avg_price_per_date.index, avg_price_per_date.values)
```

Agora vamos add uma linha de tendência no gráfico.

```python
avg_price_per_date = houses.groupby(['date'])['price'].mean()
_ = plt.title('Preços das casas ao longo do tempo')


x = avg_price_per_date.index.values.astype(np.int64) // 10**9
y = avg_price_per_date.values


coefficients = np.polyfit(x, y, 1)  #lembrei dos polinômios de 1 grau
m = coefficients[0]
b = coefficients[1]


trendline = m * x + b


plt.plot(avg_price_per_date.index, avg_price_per_date.values, label='Média de Preços')
plt.plot(avg_price_per_date.index, trendline, color='blue', label='Linha de Tendência')


plt.xlabel('Data')
plt.ylabel('Preço Médio($)')
plt.title('Preços Médios de Casas por Data')
plt.legend()

plt.show()
```

**A linha de tendência, inclinada para cima, indica que houve uma tendência geral de aumento nos preços médios das casas ao longo do período analisado.**

Apesar da tendência geral de alta, podemos observar algumas variações sazonais nos preços. Em alguns períodos do ano, os preços médios parecem ser mais altos ou mais baixos do que a média geral, indicando que fatores sazonais podem influenciar o mercado imobiliário.

Os preços médios não seguem uma trajetória perfeitamente linear. Existem flutuações e períodos de maior volatilidade, indicando que o mercado imobiliário pode ser suscetível a mudanças e incertezas.

Questão 3. Quais as regiões mais caras?

Vamos investigar quais são as 10 cidades mais caras.

```python
firist_top10 = houses.groupby('city')['price_in_k'].mean().sort_values(ascending=False).head(10)
firist_top10_cities = houses[houses['city'].isin(firist_top10.index)]
firist_top10_cities.groupby('city')['price'].count()
```

**Entretanto, será que essas cidades tem um número relevante de imóveis vendidos ou tem alguns poucos e esses são outiliers.**

```python
cities_count = houses.groupby('city')['price'].count()
cities_over_30 = cities_count[cities_count > 30] #lista de cidades com amostra suficiente
cities_over_30
```

```python
houses[houses['city'].isin(cities_over_30)].groupby('city')['price'].mean().sort_values(ascending=False).head(10)
```
